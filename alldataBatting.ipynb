```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('alldatabatting.csv')
df.head()
```

```python
df.info()
```

```python
df1=df.fillna(0.0)
df1
```

```python
#Remove unneccesary columns
df2= df.drop(columns=["Unnamed: 0"])
df2
```

```python
sns.set_style("whitegrid")
```

```python
df2.hist(figsize=(12, 6), bins=10,color='gold', edgecolor="black",)
plt.suptitle("Histogram of Batting Statistics", fontsize=14)
plt.show()
```

```python
top_10Players=df.nlargest(10,"runs")
plt.figure(figsize=(10,6))
sns.barplot(x='runs',y='Name',data=top_10Players,palette="RdBu")
plt.title("Total Runs of the top 10 Players")
plt.xlabel("Total Runs")
plt.ylabel("Name of the players")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["runs"], bins=20, kde=True,color='green')
plt.title("Distribution of Runs Scored")
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot: Runs vs Balls Faced
plt.figure(figsize=(10,6))
sns.scatterplot(x=df["balls"], y=df["runs"], hue=df["sr"], palette="inferno", size=df["sr"], sizes=(20, 200))
plt.title("Runs vs Balls Faced (Colored by Strike Rate)")
plt.xlabel("Balls Faced")
plt.ylabel("Runs Scored")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.pie([df["fours"].sum(), df["sixes"].sum()], labels=["Fours", "Sixes"], autopct="%1.1f%%", colors=["Gold", "Red"])
plt.title("Proportion of Fours vs Sixes")
plt.show()
```

```python
# Filter data for Suryakumar Yadav
player_data = df[df["Name"] == "Suryakumar Yadav"]

# Extract relevant statistics
runs = player_data["runs"].values[0]
fours = player_data["fours"].values[0]
sixes = player_data["sixes"].values[0]
sr = player_data["sr"].values[0]

# Visualization - Bar Chart
stats = ["Runs", "Fours", "Sixes", "Strike Rate"]
values = [runs, fours, sixes, sr]
plt.figure(figsize=(8, 5))
plt.bar(stats, values, color=["skyblue", "lightgreen", "red", "yellow"])
plt.xlabel("Statistics")
plt.ylabel("Values")
plt.title("Batting Performance of Suryakumar Yadav")
plt.show()
```
